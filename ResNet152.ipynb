```python
import numpy as np
xs=np.load('/content/drive/MyDrive/Papaer Self driving/train.npy')
ys=np.load('/content/drive/MyDrive/Papaer Self driving/train_labels.npy')
print(xs.shape)
print(ys.shape)
```

```python
xs = xs[0:10000]
ys = ys[0:10000]
print(xs.shape)
print(ys.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(xs,ys,test_size = 0.3, random_state = 42)
x_test, x_val, y_test, y_val = train_test_split(x_val,y_val,test_size = 0.4, random_state = 42)
```

```python
#from keras.applications import VGG16
#from keras.applications.vgg16 import VGG16
#conv_base = VGG16(weights='imagenet', include_top=False, input_shape=(66, 200, 3))

from keras.applications.resnet import ResNet152
conv_base = ResNet152(weights= 'imagenet', include_top=False, input_shape= (66,200,3))


#from keras.applications.InceptionV3 import InceptionV3
#from keras.applications import InceptionV3
#conv_base = InceptionV3(weights='imagenet',include_top=False,input_shape=(66,200,3))

#from keras.applications.densenet import DenseNet201
#conv_base = DenseNet201(weights= 'imagenet', include_top=False, input_shape= (66,200,3))


from keras import models
from keras import layers
#from keras import layers
from keras.layers import Conv2D,MaxPooling2D
from keras.layers import Activation,Dropout,Flatten,Dense
model = models.Sequential()
model.add(conv_base)
model.add(Flatten())
model.add(Dense(1164,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(100,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation = 'relu'))
```

```python
from keras import optimizers
model.compile(loss='MSE',optimizer=optimizers.Adam(lr=1e-4),metrics=['mse'])
#from sklearn.model_selection import train_test_split
batch_size = 30

train_len = len(x_train)
val_len = len(x_val)
test_len = len(x_test)
print(len(x_train), len(y_train))
print(len(x_val), len(y_val))
print(len(x_test), len(y_test))
```

```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
train_datagen = ImageDataGenerator(rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
test_datagen = ImageDataGenerator (rescale = 1./255)
history = model.fit_generator(train_datagen.flow(x_train,y_train, batch_size=batch_size, shuffle=True),
    steps_per_epoch=train_len//batch_size,
    epochs=30,
    validation_data=test_datagen.flow(x_val,y_val),
    validation_steps=val_len//batch_size,
)
```

```python
from sklearn.metrics import mean_squared_error
pred= model.predict(x_test)
print(np.sqrt(mean_squared_error(y_test,pred)))
```
